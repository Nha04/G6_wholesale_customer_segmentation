# Import All Library


What we do here:
- Import all necessary libraries (pandas, numpy, matplotlib, seaborn, scikit-learn)
- Configure plots to look professional (whitegrid style)
- Set up the environment for the entire analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
from scipy.cluster.hierarchy import dendrogram, linkage

sns.set_style("whitegrid")
%matplotlib inline

print("Libraries loaded successfully!")
```

# Load The Dataset

We start by loading the dataset. This is wholesale customer data from a real distributor. Let's take a first look on what we do here:
- Load the Wholesale Customers dataset from Google Drive
- Display first 5 rows to inspect data structure
- Check dataset info (data types, non‑null counts)
- Verify there are no missing values

```python
# Load data from local file (in the same directory as the notebook)
df = pd.read_csv("Wholesale customers data.csv")

# Show first 5 rows
print("First 5 rows:")
print(df.head())

# Show basic info
print("\nDataset info:")
print(df.info())

# Check for missing values
print("\nMissing values per column:")
print(df.isnull().sum())
```

# Exploratory Data Analysis (EDA)

In this part our main goal is to understand the data.                   
What we do here:
- Summary statistics (mean, std, min, max, quartiles)
- Drop 'Channel' and 'Region' – we only want spending data for clustering
- Plot histograms of each spending category to see distributions
- Create a correlation heatmap to find relationships between products

```python
# Summary statistics
print("Statistical summary:")
print(df.describe())

# Drop Channel and Region for clustering (we only want spending data)
X = df.drop(['Channel', 'Region'], axis=1)
print("\nFeatures used for clustering (spending categories):")
print(X.columns.tolist())

# Plot histograms of each spending category
X.hist(figsize=(12, 10), bins=30, edgecolor='black')
plt.suptitle("Distribution of Spending per Category", fontsize=16)
plt.tight_layout()
plt.show()

# Correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(X.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Between Product Categories")
plt.show()
```

# Preprocessing – Scaling

What we do here:
- Scale the data using StandardScaler (mean=0, std=1)
- Convert back to DataFrame for easier viewing
- Clustering algorithms are distance‑sensitive – scaling ensures every category has equal importance

```python
# Scale the data (very important for clustering)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Convert back to DataFrame for easier viewing
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

print("Data scaled. Mean ~0, Std ~1 for each feature.")
print(X_scaled.describe().round(2))
```

# PCA for visualization

What we do here:
- Apply PCA to reduce 6 spending dimensions to 2 principal components
- This allows us to visualize clusters in 2D later
- Check explained variance ratio to see how much information we keep

```python
# Cell 5: PCA to help us plot clusters later
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

print("Shape after PCA:", X_pca.shape)
print("Explained variance ratio:", pca.explained_variance_ratio_)
```

# First Algorithm: K-Means

What we do here:
- Find the optimal number of clusters (k) using:
  * Elbow Method (inertia)
  * Silhouette Score (higher is better)
- Choose the k that maximizes silhouette score

```python
# Find optimal k for K-Means
inertias = []
silhouette_scores = []
k_range = range(2, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_scaled, kmeans.labels_))

# Plot Elbow
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(k_range, inertias, 'bo-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method')

# Plot Silhouette
plt.subplot(1, 2, 2)
plt.plot(k_range, silhouette_scores, 'ro-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score (higher is better)')

plt.tight_layout()
plt.show()

# Choose best k (the one with highest silhouette)
best_k = k_range[silhouette_scores.index(max(silhouette_scores))]
print(f"Best number of clusters for K-Means: {best_k}")
```

## Train final K-Means model and evaluate

Some important key insights:
- Silhouette Score (closer to 1 is better) – how similar a point is to its own cluster vs. others
- Davies-Bouldin Index (lower is better) – average similarity between clusters
- Calinski-Harabasz Index (higher is better) – ratio of between‑cluster to within‑cluster variance

What we do here:
- Train K‑Means with the best k found
- Calculate all three metrics
- Store cluster labels for interpretation

```python
# Final K-Means model
final_kmeans = KMeans(n_clusters=best_k, random_state=42, n_init=10)
kmeans_labels = final_kmeans.fit_predict(X_scaled)

# Add cluster labels to original data for interpretation
X['KMeans_Cluster'] = kmeans_labels

# Calculate metrics
kmeans_sil = silhouette_score(X_scaled, kmeans_labels)
kmeans_db = davies_bouldin_score(X_scaled, kmeans_labels)
kmeans_ch = calinski_harabasz_score(X_scaled, kmeans_labels)

print(f"K-Means (k={best_k}) performance:")
print(f"  Silhouette Score: {kmeans_sil:.4f}")
print(f"  Davies-Bouldin Index: {kmeans_db:.4f}")
print(f"  Calinski-Harabasz Index: {kmeans_ch:.2f}")
```

## Visualize K-Means clusters using PCA

What we do here:
- Plot each customer in 2D using PCA coordinates
- Color points by their K‑Means cluster label
- Mark cluster centers with red X's
- This gives a visual sense of how well separated the clusters are

```python
# Plot K-Means clusters in 2D (PCA)
plt.figure(figsize=(10, 7))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=kmeans_labels, cmap='viridis', alpha=0.7)
plt.scatter(final_kmeans.cluster_centers_[:, 0], final_kmeans.cluster_centers_[:, 1],
            marker='X', s=200, c='red', label='Centroids')
plt.title(f'K-Means Clusters (k={best_k}) projected on first 2 PCA components')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.colorbar(scatter)
plt.show()
```

## Interpret K-Means clusters (average spending per cluster)

What we do here:
- Group customers by cluster and compute mean spending on each product category
- Display as a heatmap – darker colors = higher spending
- Translate these numbers into business‑friendly cluster descriptions (e.g., "Fresh Food Lovers", "Detergent & Grocery Shoppers")

```python
# Cluster profiles – what does each cluster buy?
cluster_profiles_kmeans = X.groupby('KMeans_Cluster').mean()
print("Average spending per product category for each K-Means cluster:")
print(cluster_profiles_kmeans.round(0))

# Heatmap of cluster profiles
plt.figure(figsize=(10, 6))
sns.heatmap(cluster_profiles_kmeans, annot=True, cmap='YlGnBu', fmt='.0f')
plt.title('K-Means Cluster Spending Profiles')
plt.show()
```

# Second Algorithm: Hierarchical Clustering

What we do here:
- Create a dendrogram (tree diagram) to see how customers merge
- Use Ward linkage (minimizes variance within clusters)
- Cut the dendrogram at a height to choose number of clusters

## Create dendrogram

What we do here:
- Compute linkage matrix using Ward's method
- Plot dendrogram (truncated to 30 leaves for readability)
- Add a horizontal red line as a possible cutoff point
- Use the dendrogram to visually estimate the number of clusters

```python
# Create dendrogram
linked = linkage(X_scaled, method='ward')

plt.figure(figsize=(12, 6))
dendrogram(linked, truncate_mode='lastp', p=30, leaf_rotation=90., leaf_font_size=10.)
plt.title('Dendrogram (Ward linkage)')
plt.xlabel('Customer index')
plt.ylabel('Distance')
plt.axhline(y=15, color='r', linestyle='--', label='Possible cutoff')
plt.legend()
plt.show()
```

## Find best k for Hierarchical using silhouette

What we do here:
- Loop over k from 2 to 10
- For each k, run Agglomerative Clustering
- Compute silhouette score
- Choose k with the highest silhouette

```python
# Find optimal k for Hierarchical
hier_sil_scores = []
k_range_hier = range(2, 11)

for k in k_range_hier:
    hier = AgglomerativeClustering(n_clusters=k, linkage='ward')
    hier_labels = hier.fit_predict(X_scaled)
    hier_sil_scores.append(silhouette_score(X_scaled, hier_labels))

# Plot silhouette scores
plt.figure(figsize=(8, 4))
plt.plot(k_range_hier, hier_sil_scores, 'go-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Hierarchical Clustering: Silhouette Score')
plt.show()

best_k_hier = k_range_hier[hier_sil_scores.index(max(hier_sil_scores))]
print(f"Best number of clusters for Hierarchical: {best_k_hier}")
```

## Train final Hierarchical model and evaluate

What we do here:
- Train Hierarchical clustering with best k and Ward linkage
- Compute the same three metrics (silhouette, Davies‑Bouldin, Calinski‑Harabasz)
- Add cluster labels to the dataset for interpretation

```python
# Final Hierarchical model
final_hier = AgglomerativeClustering(n_clusters=best_k_hier, linkage='ward')
hier_labels = final_hier.fit_predict(X_scaled)

# Add labels to X for interpretation
X['Hier_Cluster'] = hier_labels

# Calculate metrics
hier_sil = silhouette_score(X_scaled, hier_labels)
hier_db = davies_bouldin_score(X_scaled, hier_labels)
hier_ch = calinski_harabasz_score(X_scaled, hier_labels)

print(f"Hierarchical (k={best_k_hier}) performance:")
print(f"  Silhouette Score: {hier_sil:.4f}")
print(f"  Davies-Bouldin Index: {hier_db:.4f}")
print(f"  Calinski-Harabasz Index: {hier_ch:.2f}")
```

## Visualize Hierarchical clusters

What we do here:
- Plot PCA projection of customers
- Color points by Hierarchical cluster labels
- Compare visually with K‑Means results

```python
# Plot Hierarchical clusters
plt.figure(figsize=(10, 7))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=hier_labels, cmap='plasma', alpha=0.7)
plt.title(f'Hierarchical Clusters (k={best_k_hier}) projected on PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(scatter)
plt.show()
```

## Interpret Hierarchical clusters

What we do here:
- Calculate average spending per cluster (like we did for K‑Means)
- Create a heatmap of cluster spending profiles
- Identify which product categories define each segment

```python
# Cluster profiles for Hierarchical
cluster_profiles_hier = X.groupby('Hier_Cluster').mean()
print("Average spending per product category for each Hierarchical cluster:")
print(cluster_profiles_hier.round(0))

# Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(cluster_profiles_hier, annot=True, cmap='YlOrRd', fmt='.0f')
plt.title('Hierarchical Cluster Spending Profiles')
plt.show()
```

# Third Algorithm: DBSCAN

What we do here:
- DBSCAN finds clusters based on density, not distance from a center
- It automatically detects outliers (noise points)
- Two key parameters: eps (neighborhood radius) and min_samples (minimum points to form a dense region)

## DBSCAN parameter tuning

What we do here:
- Try a range of eps and min_samples values
- For each combination, count clusters and noise points
- Keep only configurations with ≥2 clusters and <50% noise
- Compute silhouette score for valid ones
- Select the best (eps, min_samples) pair

```python
# DBSCAN parameter tuning
from sklearn.cluster import DBSCAN

# Try different eps values (neighborhood radius) and min_samples
eps_values = [0.5, 0.8, 1.0, 1.2, 1.5, 1.8, 2.0]
min_samples_values = [3, 4, 5, 6, 8, 10]

best_score = -1
best_eps = None
best_min_samples = None
best_labels = None

for eps in eps_values:
    for min_samp in min_samples_values:
        dbscan = DBSCAN(eps=eps, min_samples=min_samp)
        labels = dbscan.fit_predict(X_scaled)
        n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
        n_noise = list(labels).count(-1)

        # Only evaluate if more than 1 cluster and less than 90% noise
        if n_clusters >= 2 and n_noise < 0.5 * len(labels):
            try:
                score = silhouette_score(X_scaled, labels)
                if score > best_score:
                    best_score = score
                    best_eps = eps
                    best_min_samples = min_samp
                    best_labels = labels
            except:
                pass

        print(f"eps={eps}, min_samples={min_samp} -> clusters={n_clusters}, noise={n_noise}")

print(f"\nBest DBSCAN: eps={best_eps}, min_samples={best_min_samples}")
if best_score != -1:
    print(f"Silhouette Score: {best_score:.4f}")
else:
    print("No valid DBSCAN configuration found. Adjust parameters.")
```

## Train final DBSCAN model

What we do here:
- Train DBSCAN with the best parameters found
- Count number of clusters (excluding noise)
- Count noise points (labeled -1)
- Calculate percentage of noise

```python
if best_eps is not None:
    final_dbscan = DBSCAN(eps=best_eps, min_samples=best_min_samples)
    dbscan_labels = final_dbscan.fit_predict(X_scaled)

    unique_labels = set(dbscan_labels)
    n_clusters_db = len([l for l in unique_labels if l != -1])
    n_noise = list(dbscan_labels).count(-1)

    print(f"DBSCAN results:")
    print(f"  Number of clusters: {n_clusters_db}")
    print(f"  Number of noise points: {n_noise}")
    print(f"  Percentage noise: {n_noise/len(dbscan_labels)*100:.1f}%")
else:
    print("DBSCAN could not find a valid configuration. Using default parameters.")
    final_dbscan = DBSCAN(eps=1.2, min_samples=5)
    dbscan_labels = final_dbscan.fit_predict(X_scaled)
    n_clusters_db = len(set(dbscan_labels)) - (1 if -1 in dbscan_labels else 0)
    n_noise = list(dbscan_labels).count(-1)
```

## Evaluate DBSCAN metrics

What we do here:
- Compute silhouette, Davies‑Bouldin, and Calinski‑Harabasz scores (if at least 2 clusters exist)
- Compare these with K‑Means and Hierarchical later

```python
if n_clusters_db >= 2:
    db_sil = silhouette_score(X_scaled, dbscan_labels)
    db_db = davies_bouldin_score(X_scaled, dbscan_labels)
    db_ch = calinski_harabasz_score(X_scaled, dbscan_labels)
    print(f"DBSCAN performance:")
    print(f"  Silhouette Score: {db_sil:.4f}")
    print(f"  Davies-Bouldin Index: {db_db:.4f}")
    print(f"  Calinski-Harabasz Index: {db_ch:.2f}")
else:
    db_sil = db_db = db_ch = None
    print("Not enough clusters to compute metrics.")
```

## Visualize DBSCAN clusters

What we do here:
- Plot PCA projection of customers
- Color points by DBSCAN cluster (noise points are a separate color)
- Note: noise points may represent unusual spending patterns (e.g., extreme high or low spenders)

```python
plt.figure(figsize=(10, 7))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=dbscan_labels, cmap='tab10', alpha=0.7)
if best_eps:
    plt.title(f'DBSCAN Clusters (eps={best_eps}, min_samples={best_min_samples})\nNoise points = {n_noise}')
else:
    plt.title(f'DBSCAN Clusters (default)\nNoise points = {n_noise}')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(scatter)
plt.show()
```

## Interpret DBSCAN clusters (exclude noise)

What we do here:
- Filter out noise points (label = -1)
- Calculate average spending per cluster (excluding noise)
- Create heatmap to understand what each dense cluster buys
- Noise points are left out of the business interpretation (they are outliers)

```python
X['DBSCAN_Cluster'] = dbscan_labels
X_dbscan_clean = X[X['DBSCAN_Cluster'] != -1].copy()
if len(X_dbscan_clean) > 0:
    cluster_profiles_db = X_dbscan_clean.groupby('DBSCAN_Cluster').mean()
    print("Average spending per product category for each DBSCAN cluster (excluding noise):")
    print(cluster_profiles_db.round(0))

    plt.figure(figsize=(10, 6))
    sns.heatmap(cluster_profiles_db, annot=True, cmap='coolwarm', fmt='.0f')
    plt.title('DBSCAN Cluster Spending Profiles')
    plt.show()
else:
    print("No non-noise clusters found.")
```

# Comparison of all three algorithms

What we do here:
- Build a side‑by‑side comparison table
- Include: number of clusters, silhouette, Davies‑Bouldin, Calinski‑Harabasz
- Create a bar chart comparing silhouette scores
- Decide which algorithm performed best overall

## Three-way comparison table

```python
comparison_data = []
comparison_data.append(['K-Means', best_k, f"{kmeans_sil:.4f}", f"{kmeans_db:.4f}", f"{kmeans_ch:.2f}"])
comparison_data.append(['Hierarchical', best_k_hier, f"{hier_sil:.4f}", f"{hier_db:.4f}", f"{hier_ch:.2f}"])

if n_clusters_db >= 2 and db_sil is not None:
    comparison_data.append(['DBSCAN', f"{n_clusters_db} (+{n_noise} noise)", f"{db_sil:.4f}", f"{db_db:.4f}", f"{db_ch:.2f}"])
else:
    comparison_data.append(['DBSCAN', f"{n_clusters_db} (+{n_noise} noise)", 'N/A', 'N/A', 'N/A'])

comparison_df = pd.DataFrame(comparison_data, columns=['Algorithm', 'Clusters', 'Silhouette', 'Davies-Bouldin', 'Calinski-Harabasz'])
print("=== THREE ALGORITHM COMPARISON ===")
print(comparison_df)
```

## Bar chart for silhouette scores

What we do here:
- Extract valid silhouette scores (ignore 'N/A')
- Create a bar chart with algorithms on x‑axis, silhouette on y‑axis
- Add value labels on top of each bar
- Higher bar = better cluster separation

```python
valid_models = comparison_df[comparison_df['Silhouette'] != 'N/A']
if len(valid_models) > 0:
    plt.figure(figsize=(8, 5))
    plt.bar(valid_models['Algorithm'], valid_models['Silhouette'].astype(float), color=['blue', 'green', 'orange'])
    plt.ylabel('Silhouette Score')
    plt.title('Silhouette Score Comparison (higher is better)')
    plt.ylim(0, 1)
    for i, v in enumerate(valid_models['Silhouette'].astype(float)):
        plt.text(i, v + 0.02, f'{v:.3f}', ha='center')
    plt.show()
```

# Final conclusion

What we do here:
- Identify the best algorithm based on silhouette score
- Translate clusters into actionable business insights
- Summarize what we learned from this project

```python
print("=== FINAL CONCLUSION ===")

best_algorithm = None
best_sil_value = -1
for idx, row in comparison_df.iterrows():
    if row['Silhouette'] != 'N/A':
        val = float(row['Silhouette'])
        if val > best_sil_value:
            best_sil_value = val
            best_algorithm = row['Algorithm']

if best_algorithm:
    print(f"Best algorithm based on Silhouette Score: {best_algorithm} with score {best_sil_value:.4f}")
else:
    print("Could not determine best algorithm.")

print("\nBusiness Insights:")
print("- DBSCAN automatically detects outliers (noise points) – these may be unusual customers.")
print("- K-Means and Hierarchical produce similar cluster structures.")
print("- Based on cluster spending profiles, we can create targeted promotions:")
print("  * Cluster A: High spenders on Fresh & Frozen -> offer fresh produce coupons")
print("  * Cluster B: High spenders on Grocery & Detergents -> bundle discounts")
print("  * Cluster C: Balanced spending -> general loyalty rewards")

print("\nProject completed successfully!")
```
